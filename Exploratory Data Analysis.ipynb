#How many Male & Female customer's are there ?

```python
plt.figure(figsize = (25,25))
plt.pie(data_01["gender"].value_counts( ), labels = data_01["gender"].value_counts().index, autopct="%1.0f%%", explode= (0.1, 0.1) , shadow=True )
plt.title("Gender Distribution", fontweight='bold')
plt.show()
```

*  **Interpretation:-**

*40 %* of customer are *Male* where as *60 %* of customer are *Female*.

# Whic age grop has the most & least customer ?

```python
age_groups = [0, 18, 25, 35, 45, 55, 65, 100]
labels = ['0-17', '18-24', '25-34', '35-44', '45-54', '55-64', '65+']
```

```python
data_01['Age Group'] = pd.cut(data_01['age'], bins=age_groups, labels=labels)
```

```python
fig, ax = plt.subplots(figsize=(25,25))
data_01['Age Group'].value_counts().sort_index().plot(kind='bar', color=mcolors.TABLEAU_COLORS, ax=ax)
ax.set_xlabel('Age Group')
ax.set_ylabel('Count')
plt.xticks(rotation=0)
ax.set_title('Customer Age Group Distribution', fontweight='bold')
for i in ax.containers:
    ax.bar_label(i)
```

*  **Interpretation:-**

*35-44* age group has the  most customer with the count of *19436* where as *0-17* age group has the  least customer with the count of *1844* .

# Which category of products is sold the most & least ?

```python
fig, ax = plt.subplots(figsize=(25,25))
data_01['category'].value_counts().plot(kind='bar', color=mcolors.TABLEAU_COLORS)
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.title('Count of Orders by Category', fontweight='bold')
for i, count in enumerate(data_01['category'].value_counts()):
    plt.text(i, count+10, str(count), ha='center', fontsize=9)
plt.show()
```

*  **Interpretation:-**

*Clothing* category of product is sold  the  most customer with the count of *34487* where as *Books* category of product is sold  the  least customer with the count of *4981*

# Which category of product has the most & least average price ?

```python
category_mean = data_01.groupby('category')['price'].mean().sort_values(ascending=False)
plt.figure(figsize=(25,25))
data_01.groupby('category')['price'].mean().sort_values(ascending=False).plot(kind='barh', color=mcolors.TABLEAU_COLORS)
plt.xlabel('Price')
plt.ylabel('Category')
plt.title('Average Price by Category', fontweight='bold')
for i, mean in enumerate(data_01.groupby('category')['price'].mean().sort_values(ascending=False)):
    plt.text(mean+0.1, i, str(round(mean, 2)), va='center', fontsize=10)
```

*  **Interpretation:-**

*Technology* category of product has the  most average with the prce of *3156.94 ₹ * where as *Food & Beverage* category of product has the  least average with the count of *15.67 ₹*

# Which payment method is often & less used ?

# Which Mall is often & less visited for shopping ?

```python
fig, ax = plt.subplots(figsize=(25,25))
data_01['shopping_mall'].value_counts().plot(kind='bar', color=mcolors.TABLEAU_COLORS)
plt.xlabel('Shopping Mall')
plt.ylabel('Count')
plt.title('Count of Customers by Shopping Mall', fontweight='bold')
for i, count in enumerate(data_01['shopping_mall'].value_counts()):
    plt.text(i, count+10, str(count), ha='center')
plt.show()
```

# What is relation between *Price* & *Quantity* ?
